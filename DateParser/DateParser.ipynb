```python
%%time
for index in df.index:
        if df.loc[index,"verbatimEventDate"] is np.nan:
            df.loc[index,"eventDate"]=np.nan
            df.loc[index,"year"]=np.nan
            df.loc[index,"month"]=np.nan
            df.loc[index,"day"]=np.nan
            df.loc[index,"impresicin"]="Empty"
        elif isinstance(df.loc[index,"verbatimEventDate"],datetime):
            df.loc[index,"eventDate"]=df.loc[index,"verbatimEventDate"].strftime("%F")
            df.loc[index,"year"]=df.loc[index,"verbatimEventDate"].strftime("%Y")
            df.loc[index,"month"]=df.loc[index,"verbatimEventDate"].strftime("X%m").replace('X0','X').replace('X','')
            df.loc[index,"day"]=df.loc[index,"verbatimEventDate"].strftime("X%d").replace('X0','X').replace('X','')
            df.loc[index,"impresicin"]="Low"
        if isinstance(df.loc[index,"verbatimEventDate"],str):
            try:
                date_element=dateparser.parse(df.loc[index,"verbatimEventDate"], settings={'PREFER_DATES_FROM': 'past', 'STRICT_PARSING': True},languages=['pt', 'es'])
                if date_element is not None: #considerar len de 4 para aos
                    date_element=dateparser.parse(df.loc[index,"verbatimEventDate"], settings={'PREFER_DATES_FROM': 'past', 'STRICT_PARSING': False},languages=['pt', 'es'])
                    df.loc[index,"eventDate"]=date_element.strftime('%F')
                    df.loc[index,"year"]=date_element.strftime("%Y")
                    df.loc[index,"month"]=date_element.strftime("X%m").replace('X0','X').replace('X','')
                    df.loc[index,"day"]=date_element.strftime("X%d").replace('X0','X').replace('X','')
                    df.loc[index,"impresicin"]="Low"
                elif date_element is None and len(df.loc[index,"verbatimEventDate"])>4: #considerar len de 4 para aos
                    date_element=dateparser.parse(df.loc[index,"verbatimEventDate"], settings={'DATE_ORDER': 'MDY','PREFER_DATES_FROM': 'past', 'STRICT_PARSING': False},languages=['pt', 'es'])
                    df.loc[index,"eventDate"]=date_element.strftime('%Y-%m')
                    df.loc[index,"year"]=date_element.strftime("%Y")
                    df.loc[index,"month"]=date_element.strftime("X%m").replace('X0','X').replace('X','')
                    df.loc[index,"day"]=np.nan
                    df.loc[index,"impresicin"]="Medium"
                elif date_element is None and len(df.loc[index,"verbatimEventDate"])<=4: #considerar len de 4 para aos
                    date_element=dateparser.parse(df.loc[index,"verbatimEventDate"], settings={'PREFER_DATES_FROM': 'past', 'STRICT_PARSING': False},languages=['pt', 'es'])
                    df.loc[index,"eventDate"]=date_element.strftime('%Y')
                    df.loc[index,"year"]=date_element.strftime("%Y")
                    df.loc[index,"month"]=np.nan
                    df.loc[index,"day"]=np.nan
                    df.loc[index,"impresicin"]="High"
            except:
                df.loc[index,"eventDate"]="check"
                df.loc[index,"year"]="check"
                df.loc[index,"month"]="check"
                df.loc[index,"day"]="check"
                df.loc[index,"impresicin"]="check"

                
```
