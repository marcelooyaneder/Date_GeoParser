# Como ingresar los valores

Como esto se basa en la georreferenciación para bases de datos en el estándar Darwin Core se agregará esta información en el campo "georeferenceRemarks" y se debe dar información de los siguientes campos mencionados anteriormente. 

- Coordinate Source
- Measurement Error
- Distance Units

La información del resto de los campos mencionados anteriormente será extraída de la misma base de datos de los campos que se muestran a continuación.

- Coordinate System -> verbatimCoordinateSystem
- Latitude -> verbatimLatitude
- Longitude -> verbatimLongitude
- Datum -> verbatimSRS 
- Precision -> coordinatePrecision

Es por esto la importancia de utilizar un vocabulario controlado. Para facilitar el uso de esta tarea se recomienda utilizar softwares como OpenRefine

## Ejemplo de ingreso de valores 

Para información más extensa del campo visitar http://rs.tdwg.org/dwc/terms/georeferenceRemarks

Ejemplos prácticos

- locality description, 50.865, km
- Assumed distance by road (Hwy. 101) | GPS, 100, m

Es importante que si ya hay anotaciones presentes en nuestra base de datos agregar el caracter "|" de separador y que esta sea la ultima información proporcionada.

**Si es que no conoces la información de Coordinate Source utiliza locality description"**

```python
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
```

```python
driver = webdriver.Chrome(executable_path=r'C:\Users\marcelo\Documents\chromedriver_win32\chromedriver.exe')
driver.get("https://georeferencing.org/georefcalculator/gci3/source/gci3.html")
```

```python
class seleniumGeorref:

    #Controlled vocabulary     
    CoordSystem_dic={   
    "decimal degrees": 0,
    "degrees minutes seconds": 2,
    "degrees decimal minutes": 1
    }
    
    #Constant
    ChoiceModel= driver.find_element_by_id("ChoiceModel")
    ChoiceCoordSource=driver.find_element_by_id("ChoiceCoordSource")
    ChoiceCoordSystem=driver.find_element_by_id("ChoiceCoordSystem")
    ChoiceDatum=driver.find_element_by_id("ChoiceDatum")
    ChoiceLatPrecision=driver.find_element_by_id("ChoiceLatPrecision")
    deg_pattern="(.*?)\°"
    min_pattern="\°(.*?)\'"
    sec_pattern="\'(.*?)\""
    
    #variables
    txtT7Lat_DegMM=driver.find_element_by_id("txtT7Lat_DegMM")
    txtT7Lat_MinMM=driver.find_element_by_id("txtT7Lat_MinMM")
    txtT7Long_DegMM=driver.find_element_by_id("txtT7Long_DegMM")
    txtT7Long_MinMM=driver.find_element_by_id("txtT7Long_MinMM")
    txtT2Dec_Lat=driver.find_element_by_id("txtT2Dec_Lat")
    txtT2Dec_Long=driver.find_element_by_id("txtT2Dec_Long")
    textFieldMeasurementError=driver.find_element_by_id("TextFieldMeasurementError")
    ChoiceDistUnits=driver.find_element_by_id("ChoiceDistUnits")
    ChoiceLongDirMM=driver.find_element_by_id("ChoiceLongDirMM")
    ChoiceLatDirMM=driver.find_element_by_id("ChoiceLatDirMM")
    txtT7Lat_DegDMS=driver.find_element_by_id("txtT7Lat_DegDMS")
    txtT7Lat_MinDMS=driver.find_element_by_id("txtT7Lat_MinDMS")
    txtT7Lat_Sec=driver.find_element_by_id("txtT7Lat_Sec")
    ChoiceLatDirDMS=driver.find_element_by_id("ChoiceLatDirDMS")
    txtT7Long_DegDMS=driver.find_element_by_id("txtT7Long_DegDMS")
    txtT7Long_MinDMS=driver.find_element_by_id("txtT7Long_MinDMS")
    txtT7Long_Sec=driver.find_element_by_id("txtT7Long_Sec")
    ChoiceLongDirDMS=driver.find_element_by_id("ChoiceLongDirDMS")

    def __init__(self,CoordSystem,verbatimLatitude,verbatimLongitude):
        self.CoordSystem=CoordSystem    
        self.verbatimLatitude=verbatimLatitude
        self.verbatimLongitude=verbatimLongitude
        
    
    def coordinatesOnly(self):
        Select(seleniumGeorref.ChoiceModel).select_by_index(1)
        if self.CoordSystem == "decimal degrees":
            Select(seleniumGeorref.ChoiceCoordSystem).select_by_index(seleniumGeorref.CoordSystem_dic[self.CoordSystem])
            #clear fields
            seleniumGeorref.txtT2Dec_Lat.clear()
            seleniumGeorref.txtT2Dec_Long.clear()
            #send keys to fields
            seleniumGeorref.txtT2Dec_Lat.send_keys(self.verbatimLatitude)
            seleniumGeorref.txtT2Dec_Long.send_keys(self.verbatimLongitude)
            
        elif self.CoordSystem == "degrees minutes seconds":
            Select(seleniumGeorref.ChoiceCoordSystem).select_by_index(seleniumGeorref.CoordSystem_dic[self.CoordSystem])
            #clear fields
            seleniumGeorref.txtT7Lat_DegDMS.clear()
            seleniumGeorref.txtT7Lat_MinDMS.clear()
            seleniumGeorref.txtT7Lat_Sec.clear()
            seleniumGeorref.txtT7Long_DegDMS.clear()
            seleniumGeorref.txtT7Long_MinDMS.clear()
            seleniumGeorref.txtT7Long_Sec.clear()
            #send keys to fields
            seleniumGeorref.txtT7Lat_DegDMS.send_keys(re.search(seleniumGeorref.deg_pattern, self.verbatimLatitude).group(1))
            seleniumGeorref.txtT7Lat_MinDMS.send_keys(re.search(seleniumGeorref.min_pattern, self.verbatimLatitude).group(1))
            seleniumGeorref.txtT7Lat_Sec.send_keys(re.search(seleniumGeorref.sec_pattern, self.verbatimLatitude).group(1))
            seleniumGeorref.txtT7Long_DegDMS.send_keys(re.search(seleniumGeorref.deg_pattern, self.verbatimLongitude).group(1))
            seleniumGeorref.txtT7Long_MinDMS.send_keys(re.search(seleniumGeorref.min_pattern, self.verbatimLongitude).group(1))
            seleniumGeorref.txtT7Long_Sec.send_keys(re.search(seleniumGeorref.sec_pattern, self.verbatimLongitude).group(1))
            #select index
            if "N" in self.verbatimLatitude.upper():
                Select(seleniumGeorref.ChoiceLatDirDMS).select_by_index(0)
            else: Select(seleniumGeorref.ChoiceLatDirDMS).select_by_index(1)
            if "E" not in self.verbatimLongitude.upper():
                Select(seleniumGeorref.ChoiceLongDirDMS).select_by_index(0)
            else: Select(seleniumGeorref.ChoiceLongDirDMS).select_by_index(1)
            
            
        elif self.CoordSystem == "degrees decimal minutes":
            Select(seleniumGeorref.ChoiceCoordSystem).select_by_index(seleniumGeorref.CoordSystem_dic[self.CoordSystem])
            #clear fields
            seleniumGeorref.txtT7Lat_DegMM.clear()
            seleniumGeorref.txtT7Lat_MinMM.clear()
            seleniumGeorref.txtT7Long_DegMM.clear()
            seleniumGeorref.txtT7Long_MinMM.clear()
            #send keys to fields
            seleniumGeorref.txtT7Lat_DegMM.send_keys(re.search(seleniumGeorref.deg_pattern, self.verbatimLatitude).group(1))
            seleniumGeorref.txtT7Lat_MinMM.send_keys(re.search(seleniumGeorref.min_pattern, self.verbatimLatitude).group(1))
            seleniumGeorref.txtT7Long_DegMM.send_keys(re.search(seleniumGeorref.deg_pattern, self.verbatimLongitude).group(1))
            seleniumGeorref.txtT7Long_MinMM.send_keys(re.search(seleniumGeorref.min_pattern, self.verbatimLongitude).group(1))
            #select index
            if "N" in self.verbatimLatitude.upper():
                Select(seleniumGeorref.ChoiceLatDirMM).select_by_index(0)
            else: Select(seleniumGeorref.ChoiceLatDirMM).select_by_index(1)
            if "E" not in self.verbatimLongitude.upper():
                Select(seleniumGeorref.ChoiceLongDirMM).select_by_index(0)
            else: Select(seleniumGeorref.ChoiceLongDirMM).select_by_index(1)
                
                
            
            
            
            
    
    
    def geographicFeature():
        Select(seleniumGeorref.ChoiceModel).select_by_index(1)
        Select(seleniumGeorref.ChoiceCoordSystem).select_by_index(0)


    
   
```
